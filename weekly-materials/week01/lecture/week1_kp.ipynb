### Where do we write and run code?

- Here, we use **Jupyter Notebooks**, a software that runs in your browser and allows you write, run, and view the outputs of your code -- along with normal text, like a word document.
- You can access Jupyter by going to <https://jupyter.utoronto.ca> (see the JupyterHub guide on the course website).
